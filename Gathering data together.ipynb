```python
import json
from matplotlib import pyplot
%matplotlib inline
```

```python
tags = json.load(open('by-tags.txt'))
```

```python
tags['IT'][:2]
```

```python
import hashlib
```

```python
uniques = {}
for i in tags:
    for j in tags[i]:
        key = hashlib.md5(j['text'].encode('utf-8')).hexdigest()
        if key in uniques:
            uniques[key]['tag'] += [i]
        else:
            uniques[key] = {'tag': [i], 'text': j['text'], 'retweets': j['retweets'], 'likes': j['likes'], 'userId': j['userId']}
```

```python
json.dump(uniques, open("uniques.json", 'w'))
```

```python
import collections
collections.Counter([uniques[i]['userId'] for i in uniques])
```

```python
json.dump(collections.Counter([uniques[i]['userId'] for i in uniques]), open("best.json", 'w'))
```
